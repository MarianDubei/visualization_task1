```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
csv_filename = "GLB.Ts_dSST.csv"
df = pd.read_csv(csv_filename, skiprows=1)
```

```python
df.head()
```

```python
sns.set(rc={'figure.figsize':(18,10)})
sns.set_style({'axes.facecolor':'white',
               'grid.color': '#e8e8e8',
               'xtick.color':'#999999','ytick.color':'#999999'
              })
# J-D is average per year
ax = sns.scatterplot(x=df["Year"],
                y=df["J-D"],
                color='#f26c08',
                edgecolor='black',
                linewidth=0.5,
                s=100
               )
plt.text(1883, 0.9, "Annual GLobal Surface Temperature,\nRelative to Late 19th Century Average", fontfamily="Lucida Sans Unicode", fontsize=14)
ax.set(xlabel=None, ylabel=None)
ylabels = list(map(lambda x: "+"+'{:,.1f}'.format(x)+"$^\circ$" if x > 0 else '{:,.1f}'.format(x)+"$^\circ$", ax.get_yticks()))
ylabels[-2] += "C"
ax.set_yticklabels(ylabels)
ax.set_xticks(range(1880, 2020, 10))

ind = ylabels.index('-0.2$^\\circ$')
gridlines = ax.yaxis.get_gridlines()
gridlines[ind].set_color("#989898")

print()
l_point_years = [1904, 1998]
r_point_years = [1944, 2014, 2015, 2017]
for year in l_point_years:
    x = year
    y = df.loc[df['Year'] == year]["J-D"]
#     plt.annotate(str(year)+" ", xy=(x, y), xytext=(3,0.5))
    plt.text(x-6, y-0.015, str(year))
for year in r_point_years:
    x = year
    y = df.loc[df['Year'] == year]["J-D"]
    plt.text(x+1.5, y-0.013, str(year))
plt.text(2016+1.5, df.loc[df['Year'] == 2016]["J-D"]-0.013, "2016", weight='bold')
    
plt.text(2010, -0.2, "HOTTER THAN THE\n1880-1899 AVERAGE", fontfamily="Tahoma", fontsize=14)
plt.text(2016.5, -0.25, "COLDER", fontfamily="Lucida Sans Unicode", fontsize=14)
plt.show()
```
